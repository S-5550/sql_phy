# select weather,AVG(temperature) from dataset_1 group by weather;¶

```python
sql.groupby('weather')['temperature'].mean().reset_index().rename(columns={'temperature':'avg_count'})
```

# select weather,count(temperature) from dataset_1 group by weather;¶

```python
sql.groupby('weather')['temperature'].count().reset_index().rename(columns={'temperature':'count_temp'})
```

# select weather,count(distinct temperature) from dataset_1 group by weather;¶

```python
sql.groupby('weather')['temperature'].nunique().reset_index().rename(columns={'temperature' : 'count_distinct_temp'})
```

# select weather,SUM(temperature) as sum_temp from dataset_1 group by weather;¶

```python
sql.groupby('weather')['temperature'].sum().reset_index().rename(columns={'temperature':'sum_temp'})
```

# select weather,min(temperature) as min_temp from dataset_1 group by weather;

```python
sql.groupby('weather')['temperature'].min().reset_index().rename(columns={'temperature':'min_temp'})
```

# select weather,max(temperature) as max_temp from dataset_1 group by weather;¶

```python
sql.groupby('weather')['temperature'].max().reset_index().rename(columns={'temperature':'max_temp'})
```
